```python
import openpyxl
from collections import defaultdict

# Load the workbook and select the active worksheet
workbook = openpyxl.load_workbook('/content/drive/MyDrive/Colab_Notebooks/GNN/GraphsClassification - PPIGraphNet/data/raw_data/idmapping_2024_06_24.xlsx')
worksheet = workbook.active

# Initialize an empty dictionary with lists as default values
result_dict = defaultdict(list)
unique_values_1 = set()

# Iterate over the rows in the worksheet
for row in worksheet.iter_rows(min_row=2, values_only=True):
    key = row[0]    # First column
    value = row[1]  # Second column
    result_dict[key].append(value)  # Append the value to the list of the corresponding key
    unique_values_1.add(value)

# Print the result dictionary
print(len(result_dict))
print(len(unique_values_1))

for key, value in result_dict.items():
    print(key, len(value), value)
```

```python
pd_pdb = pd.DataFrame(columns=['protein1_pdb', 'protein2_pdb', 'protein1_uniprot', 'protein2_uniprot', 'interaction_type'])

for index, row in csv_file.iterrows():
    col1_value = row['protein1_uniprot']
    col2_value = row['protein2_uniprot']
    interaction_type_value = row['interaction_type']
    print(index, col1_value, col2_value)

    if col1_value not in result_dict or col2_value not in result_dict:
        print('Skipping')
        continue

    l1 = result_dict[col1_value]
    l2 = result_dict[col2_value]
    pairs = []

    for i in l1:
        for j in l2:
            pairs.append({
                'protein1_pdb': i,
                'protein2_pdb': j,
                'protein1_uniprot': col1_value,
                'protein2_uniprot': col2_value,
                'interaction_type': interaction_type_value
            })
    pd_pdb = pd.concat([pd_pdb, pd.DataFrame(pairs)], ignore_index=True)

print(pd_pdb.shape)
pd_pdb.to_csv('/content/drive/MyDrive/Colab_Notebooks/GNN/GraphsClassification - PPIGraphNet/data/processed_csv/interaction_pdb_db.csv', index=False)
```

```python
import urllib.parse
import urllib.request
import pandas as pd
import numpy as np

file_path = '/content/drive/MyDrive/Colab_Notebooks/GNN/GraphsClassification - PPIGraphNet/data/processed_csv/interaction_pdb_db.csv'
chunksize = 10**6  # Process 1 million rows at a time

# Initialize an empty list to store unique rows
unique_rows = []

for chunk in pd.read_csv(file_path, chunksize=chunksize):
    # Normalize pairs by sorting protein1_pdb and protein2_pdb
    chunk['pair'] = chunk.apply(lambda row: tuple(sorted((row['protein1_pdb'], row['protein2_pdb']))), axis=1)

    # Drop rows where 'protein2_uniprot' equals 'protein1_uniprot'
    chunk = chunk[chunk['protein2_uniprot'] != chunk['protein1_uniprot']]

    # Drop duplicates within the current chunk
    chunk_unique = chunk.drop_duplicates(subset=['pair', 'interaction_type'])

    # Append unique rows to the list
    unique_rows.append(chunk_unique)

# Concatenate all unique rows
unique_df = pd.concat(unique_rows)

# Drop duplicates across all chunks
final_df = unique_df.drop_duplicates(subset=['pair', 'interaction_type']).drop(columns='pair')

# Save the final DataFrame to a new CSV file
output_path = '/content/drive/MyDrive/Colab_Notebooks/GNN/GraphsClassification - PPIGraphNet/data/processed_csv/interaction_pdb_db_uniquepairs.csv'
final_df.to_csv(output_path, index=False)
print(f"Processed file saved to: {output_path}")
```
